```python
from keras.preprocessing.image import ImageDataGenerator
from skimage import io
```

```python
datagen = ImageDataGenerator(
                    rotation_range=30,
                    width_shift_range=0.1,
                    height_shift_range=0.1,
                    shear_range=0.1,
                    zoom_range=0.1,
                    fill_mode='nearest',
                    cval=255
                    )
```

```python
import numpy as np
import os
from PIL import Image
```

```python
img_dir = '/content/drive/MyDrive/data/img/parkinson/'
dataset = []
par_images = os.listdir(img_dir)
for i, name in enumerate(par_images):
      if(name.split('.')[1] == 'png'):
          image = io.imread(img_dir + name)
          dataset.append(np.array(image))
x = np.array(dataset)
print(x.shape)

i = 0
for batch in datagen.flow(x, batch_size=16,
                          save_to_dir='/content/drive/MyDrive/parkinson_augmented',
                          save_prefix='aug',
                          save_format='png'):
      i += 1
      if i > 15:
          break
```
